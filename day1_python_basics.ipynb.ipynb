```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print("Factorial of 5 is:", factorial(5))
```

```python
#Reversing a string using a function and just for loop
def reverse_string(s):
    reversed_s=""
    for i in s:
        reversed_s=i+reversed_s
    return reversed_s
s="Eeswar"
print("The reverse of string ",s," is ",reverse_string(s))
```

```python
##Sum of all the numbers in a range
start,end=map(int, input().split())
sum_total=0
sum_even=0
for i in range(start, end+1):
    sum_total+=i
    if i%2==0:
        sum_even+=i
print("The total sum of all the elements from ", start,"to ", end,"is ", sum_total)
print("The total sum of all the even elements from ",start, "to ",end,"is ",sum_even)
print("The total sum of all the odd elements from ",start, "to ",end,"is ",sum_total-sum_even)
```
